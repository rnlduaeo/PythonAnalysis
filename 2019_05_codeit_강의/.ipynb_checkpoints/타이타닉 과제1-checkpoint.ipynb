```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv', index_col=0)
df.head()
```

```python
sns.distplot(df['Age'], bins=20) #승객은 20-30대가 제일 많다
```

```python
df[['Fare', 'Age']].sort_values(by='Fare', ascending=False) #가장 높은 요금을 낸 사람은 30대이다
```

```python
df.plot(kind='scatter', x='Fare', y='Age') #가장 높은 요금을 낸 사람은 30대이다.
```

```python
df['Survived'].value_counts().plot(kind='pie') #사망자가 생존자 보다 많다
```

```python
df['Pclass'].value_counts() #3등실을 제일 많이 탔다
```

```python
sns.kdeplot(df['Survived'], df['Pclass']) #3등급은 0 에 밀집, 2등급은 비슷, 1등급은 1에 밀집 즉 1등급이 제일 생존률이 높다
```

```python
sns.stripplot(data=df, x='Survived', y='Age')
```

```python
sns.violinplot(data=df, x='Survived', y='Age') #나이가 어릴 수록 생존률이 높다? 나이랑은 별로 상관이 없는 듯 하다. 
```

```python
sns.stripplot(data=df, x='Survived', y='Age', hue='Sex') #확실히 여자가 생존률이 높다.. 이렇게 여러 카테고리로 생존률을 비교하고 싶다면 Stripplot을 사용하고 hue를 통해 색깔로 구분 ㅋㅋ
```

```python
df.loc[df['Pclass']==3, 'Survived'].value_counts().plot(kind='pie')
```

```python
sns.distplot(df.loc[df['Survived']==1, 'Age'], bins=10) #나이가 어릴 수록 생존률이 높다?
```

```python
sns.countplot(data = df.loc[df['Survived']==1, :], x='Sex')
```

```python
sns.countplot(data = df.loc[df['Survived']==0, :], x='Sex')
```
