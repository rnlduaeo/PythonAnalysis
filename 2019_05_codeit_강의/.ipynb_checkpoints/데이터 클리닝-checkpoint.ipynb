```python
df=pd.read_csv('data/beer.csv')
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.25)
```

```python
df['abv'].quantile(0.75)
```

```python
q1 = df['abv'].quantile(0.25)
q3 = df['abv'].quantile(0.75)
iqr = q3-q1
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 * iqr)
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055 #결측값 변경
df[condition]
```

```python
df.drop(df[condition].index, inplace=True) #맥주가 아니기 때문에 그냥 제거
```

```python
df[condition]
```

```python
df.plot(kind='box', y='abv')
```

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/movie_metadata.csv')

#너무 예산이 큰 일부 영화를 제거하고 다시 분석
#예산을 기준으로 75% 지점에서 5 Iqr 만큼 더한 것보다 큰 예산의 영화는 제거
q1 = df['budget'].quantile(0.25)
q3 = df['budget'].quantile(0.75)
iqr = q3-q1
#아래 부분 중 5를 마음대로 조정하면서 이상점 제거가능
#예를 들어 10으로 두면 더 적은 이상점 제거하는 것 
con = df['budget'] > (q3 + 5*iqr)
df.drop(df[con].index, inplace=True)
df.plot(kind='scatter', x='budget', y='imdb_score')
```

```python
%matplotlib inline
import pandas as pd

#상위 15개 제거하고 산점도 다시 그리기
df = pd.read_csv('data/movie_metadata.csv')
sort = df['budget'].sort_values(ascending=False)
#sort.head(10)

#iloc 말고 sort.head(15)로 바꿔도 됨...sort.head(15).index
df.drop(sort.iloc[:15].index, inplace=True)

df.plot(kind='scatter', x='budget', y='imdb_score')
# 코드를 작성하세요.
```
