-- 2,000명의 코드잇 대학교 학생들이 수강신청을 했습니다.

수강신청에는 다음 3개의 조건이 있습니다.

“information technology” 과목은 심화과목이라 1학년은 수강할 수 없습니다.
“commerce” 과목은 기초과목이고 많은 학생들이 듣는 수업이라 4학년은 수강할 수 없습니다.
수강생이 5명이 되지 않으면 강의는 폐강되어 수강할 수 없습니다.
기존 DataFrame에 “status”라는 이름의 column을 추가하고, 학생이 수강 가능한 상태이면 “allowed”, 수강 불가능한 상태이면 “not allowed”를 넣어주세요.

```python
import pandas as pd

df = pd.read_csv('enrolment_1.csv')

df['status'] = 'allowed'

con1 = (df['course name'] == 'information technololgy') & (df['year'] == 1)
con2 = (df['course name'] == 'commerce') & (df['year'] == 4)
df.loc[con1 | con2, 'status'] = 'not allowed'

allowed = df.loc[df['status'] == 'allowed', 'course name']

list = allowed.value_counts()

not_allowed = list[list<5].index

for i in not_allowed:
    df.loc[df['course name'] == i, 'status'] = 'not allowed' # if를 안넣어줘도 dataframe안에 condition으로 넣어줘도 됨.
df
```

```python
import pandas as pd

df = pd.read_csv('enrolment_2.csv')

# 코드를 작성하세요.
df['room assignment'] = 'not assigned'
allowed = df[df['status'] == 'allowed']
enrol_cnt = allowed['course name'].value_counts()
audi = enrol_cnt[enrol_cnt>=80].index
large = enrol_cnt[(enrol_cnt<80) & (enrol_cnt>=40)].index
medium = enrol_cnt[(enrol_cnt<40) & (enrol_cnt>=15)].index
small = enrol_cnt[(enrol_cnt<15) & (enrol_cnt>=5)].index

for i in audi:
    df.loc[df['course name'] == i, 'room assignment'] = 'Auditorium'

for i in large:
    df.loc[df['course name'] == i, 'room assignment'] = 'Large room'

for i in medium:
    df.loc[df['course name'] == i, 'room assignment'] = 'Medium room'
    
for i in small:
    df.loc[df['course name'] == i, 'room assignment'] = 'Small room'
# 정답 출력
df.loc[df['status'] == 'not allowed', 'room assignment'] = 'not assigned'

df
```

```python
import pandas as pd

df = pd.read_csv('enrolment_3.csv')

# 코드를 작성하세요.
sort = df.sort_values(by='course name')
assigned = sort['room assignment'] != 'not assigned'
audi = sort[(sort['room assignment'] == 'Auditorium') & assigned]
small = sort[(sort['room assignment'] == 'Small room') & assigned]
medium = sort[(sort['room assignment'] == 'Medium room') & assigned]
large = sort[(sort['room assignment'] == 'Large room') & assigned]

audi_un = pd.unique(audi['course name'])
small_un = pd.unique(small['course name'])
medium_un = pd.unique(medium['course name'])
large_un = pd.unique(large['course name'])

cnt = 1
for i in audi_un:
    df.loc[df['course name'] == i, 'room assignment'] = "Auditorium-" + str(cnt)
    cnt=cnt + 1

cnt = 1
for i in small_un:
    df.loc[df['course name'] == i, 'room assignment'] = "Small-" + str(cnt)
    cnt=cnt + 1

cnt = 1
for i in medium_un:
    df.loc[df['course name'] == i, 'room assignment'] = "Medium-" + str(cnt)
    cnt=cnt + 1

cnt = 1
for i in large_un:
    df.loc[df['course name'] == i, 'room assignment'] = "Large-" + str(cnt)
    cnt=cnt + 1


df.rename(columns = {'room assignment' : 'room number'}, inplace=True)
df.loc[df['status'] == 'not allowed', 'room number'] = 'not assigned' 

df
# 정답 출력
#df
```
