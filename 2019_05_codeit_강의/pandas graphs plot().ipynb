```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('silicon_valley_details.csv')
not_zero=df.loc[df['count']!=0]
is_Adobe = not_zero['company'] == 'Adobe'
adobe = not_zero.loc[is_Adobe]
adobe2 = adobe[adobe['race'] == 'Overall_totals']
adobe3 = adobe2[['job_category', 'count']]
con1 = (adobe3['job_category']!='Totals') & (adobe3['job_category']!='Previous_totals')

adobe4 = adobe3.loc[con1].set_index('job_category')
adobe4.plot(kind='pie', y='count')
```
