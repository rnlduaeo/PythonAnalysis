```python
import requests
import numpy as np

rating_pages = [];
for i in range(2010, 2019):
    for j in range(1, 13):
        for k in range (0, 5):
            page = requests.get("https://workey.codeit.kr/ratings/index?year={0}&month={1}&weekIndex={2}".format(i, j, k));
            rating_pages.append(page.text);

print(len(rating_pages)) # 가져온 총 페이지 수 
print(rating_pages[0]) # 첫 번째 페이지의 HTML 코드
```
