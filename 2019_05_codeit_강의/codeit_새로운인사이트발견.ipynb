```python
import pandas as pd

df = pd.read_csv('data/occupations.csv')
#groupby 에서 mean은 숫자값을 지니는 컬럼에서만 작동한다. 따라서 범주형 값의 비율을 알기 위해서는 수치형으로 바꾸고 mean()함수를 활용하면 된다.
mapping = {
    'F' : 1,
    'M' : 0
}
df['gender'] = df['gender'].map(mapping)
occ_group = df.groupby('occupation')
occ_group.mean()['gender'].sort_values(ascending=False)
# 코드를 작성하세요.
```
